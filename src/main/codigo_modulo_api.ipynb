```python
Spain_food = df.loc[:,['Area', 'Item','Y2007', "Y2008",'Y2012']]
Spain_food = Spain_food[Spain_food.Area == 'Spain']
Spain_food
```

```python
Spain_production.to_json()
```

```python
Spain_production.to_json("Spain_production.json")
```
